# Assignment 1- Data Visualization

This code creates a variety plots and graphs of the "titanic" data set using the "ggplot2" package. Multiple variables and styles of graphs are looked at. Not all the graphs show/represent important findings. The main purpose of this assignment was to get familiar with data visualization options with "ggplot2".

### Uploading Data & Cleaning

```r
setwd("C:/Users/student/Documents/Data Analysis Practice Data Sets/Titanic Practice")
getwd()
titanic = read.csv("train.csv")
titanic$Survived = factor(titanic$Survived)
titanic$Pclass = factor(titanic$Pclass)
library(ggplot2)
```

### Males versus Females Survival Rate Bar Chart

```r
ggplot(titanic) +
geom_bar(mapping = aes(x=Sex, fill = Survived)) +
scale_fill_discrete(breaks=c("0", "1"),
                     labels=c("Died", "Survived"))
```

### Fare and Wage on Survival Scatter Plot

```r
ggplot(titanic) +
geom_point(mapping = aes(x=Fare, y=Age, color=Survived), alpha = 0.8) +
scale_color_discrete(breaks=c("0", "1"),
                    labels=c("Died", "Survived"))
```

### Age and Survivor Box Plot

```r
ggplot(titanic) +
geom_boxplot(mapping = aes(x=Survived, y=Age), fill = "lightblue") +
scale_x_discrete(breaks=c("0", "1"),
                    labels=c("Died", "Survived"))
```

### Survivors by Ticket Class

```r

ggplot(titanic) +
    geom_bar(mapping = aes(x=Survived), fill = "purple") +
    facet_grid(. ~ Pclass) +
    scale_x_discrete(breaks=c("0", "1"),
                        labels=c("Died", "Survived")) +
    theme(panel.background = element_rect(fill = "light blue"))
```

### Port of Embarkation and Surviving Polar Graph

```r
ggplot(titanic) +
    geom_bar(mapping = aes(x=Embarked, fill=Survived)) +
    scale_fill_discrete(breaks=c("1", "0"),
                        labels=c("Survived", "Died")) +     
    scale_x_discrete(breaks=c("S", "Q", "C"),
                        labels=c("Southampton", "Queenstown", "Cherbourg")) +
    scale_fill_brewer(palette = "Greens") +
    coord_polar()
```

### Average Ticket Price per Class

```r
ggplot(titanic, aes(x=Pclass, y=Fare)) +
    geom_bar(stat="summary",fun.y="mean", aes(fill=Pclass)) +
    scale_x_discrete(name = "Ticket Class") +
    scale_fill_discrete(guide=FALSE) +
    theme_classic()
```
